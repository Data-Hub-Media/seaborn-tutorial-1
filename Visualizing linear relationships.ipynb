```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(color_codes=True)
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.shape
```

```python
tips.info()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.regplot(data=tips, x="total_bill", y="tip")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", 
           hue="smoker", col="smoker", row="time",
           truncate=False, ci=None)
```

```python
sns.lmplot(x="size", y="tip", data=tips, truncate=False, x_jitter=.05)
```

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean, truncate=False)
```
