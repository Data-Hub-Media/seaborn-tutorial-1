## 단변수
* 변수가 하나일 때
* 단변수(Univariable) - 1변수(독립변수가 하나)
* 단변량(Univariate) - 1변량(종속변수가 하나)
* 출처 : [독립변수와 종속변수 - 위키백과, 우리 모두의 백과사전](https://ko.wikipedia.org/wiki/%EB%8F%85%EB%A6%BD%EB%B3%80%EC%88%98%EC%99%80_%EC%A2%85%EC%86%8D%EB%B3%80%EC%88%98)
    * 실험이나 모델링에서 사용되는 변수는 독립 변수(independent variable), 종속 변수(dependent variable) 혹은 기타로 분류할 수 있다. 독립 변수는 입력값이나 원인을 나타내며, 종속 변수는 결과물이나 효과를 나타낸다. 
    * 기계 학습 혹은 패턴 인식에서는 특징 (feature) 이라고도 한다. 그냥 단순히 입력 변수(input variable)이라고도 한다.
    * 독립 변수는 설명 변수(Explanatory variable)라고 하기도 한다. 독립 변수로 취급하는 것의 수량이 통계적으로 독립적이지 않을 경우에 한하여 더 선호된다.

```python
# 라이브러리를 로드합니다.
import pandas as pd
import seaborn as sns
```

```python
# tips 데이터셋을 가져옵니다.

tips = sns.load_dataset("tips")
tips.shape
```

## distplot

```python
sns.distplot(tips["tip"])
```

## boxplot

```python
# boxplot 으로 시각화 합니다.
sns.distplot(tips["tip"])
sns.boxplot(tips["tip"])
```

## boxenplot

```python
sns.boxenplot(tips["tip"])
```

## violinplot

```python
sns.violinplot(tips["tip"])
```

## subplots
* distplot, boxplot, boxenplot, violinplot, swarmplot 을 그립니다.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(15, 8))

sns.distplot(tips["tip"], ax=axes[0])
sns.boxplot(tips["tip"], ax=axes[1])
sns.boxenplot(tips["tip"], ax=axes[2])
sns.violinplot(tips["tip"], ax=axes[3])
sns.swarmplot(tips["tip"], ax=axes[4])
```

## 이상치(outlier) 구하기
* https://ko.wikipedia.org/wiki/%EC%83%81%EC%9E%90_%EC%88%98%EC%97%BC_%EA%B7%B8%EB%A6%BC
* IQR = Q3 - Q1
* outlier
    * Q3 + 1.5(IQR)
    * Q1 - 1.5(IQR)

```python
desc = tips["tip"].describe()
desc
```

```python
IQR = desc["75%"] - desc["25%"]
IQR_15 = IQR * 1.5
IQR_15
```

```python
desc["75%"] + IQR_15
```

```python
desc["25%"] - IQR_15
```
