## Line plots on multiple facets
* https://seaborn.pydata.org/examples/faceted_lineplot.html

```python
import seaborn as sns
sns.set(style="ticks")
%matplotlib inline
```

```python
dots = sns.load_dataset("dots")
dots.shape
```

```python
dots.head()
```

```python
dots.info()
```

```python
dots.coherence.unique()
```

```python
dots.coherence.nunique()
```

```python
sns.palplot(sns.color_palette("rocket_r", 6))
```

```python
# Define a palette to ensure that colors will be
# shared across the facets
palette = dict(zip(dots.coherence.unique(),
                   sns.color_palette("rocket_r", 6)))
palette
```

```python
sns.relplot(x="time", y="firing_rate", data=dots, 
            col="align", size="choice", size_order=["T1", "T2"], 
            hue="coherence", palette=palette, height=5, aspect=.75,  legend="full", 
            facet_kws=dict(sharex=False),
            kind='line')
```

```python
# Plot the lines on two facets
sns.relplot(x="time", y="firing_rate",
            hue="coherence", size="choice", col="align",
            size_order=["T1", "T2"], palette=palette,
            height=5, aspect=.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=dots)
```
